```python
import pandas as pd
```

```python
with open('nanoseconds.txt', 'r') as f:
    data = f.read()

nanoseconds = []
for nanoseconds_literal in data.split("\n"):
    try:
        nanoseconds.append(float(nanoseconds_literal))
    except ValueError:
        print(f"{nanoseconds_literal} is not a float")
```

```python
# Create a dataframe for all the value below
cpp_O3 = nanoseconds[:6]
cpp_O1 = nanoseconds[6:12]
cpp_intrinsic = nanoseconds[12:18]
python = nanoseconds[18::2]
python_numpy = nanoseconds[19::2]

df = pd.DataFrame({
    'C++ O3': cpp_O3,
    'C++ O1': cpp_O1,
    'C++ intrinsic': cpp_intrinsic,
    'Python': python,
    'Python Numpy': python_numpy
}, index=['1000', '10k', '100k', '1M', '10M', '100M'])

# Plot line graph for the dataframe
df.plot(figsize=(16, 8), kind='line')
```
